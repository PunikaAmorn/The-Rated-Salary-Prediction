```python
def accuracy_range(train,test):
    num_all = len(test)
    num_correct = 0
    if len(train)==len(test):
        for i,v in enumerate(train):
            if v == test[i]:
                num_correct+=1
                continue
            if test[i] <1.5 and v==1.5:
                num_correct+=1
            elif test[i] >1.5000 and test[i] < 2.0001 and v==1.75005:
                num_correct+=1
            elif test[i] >2.0000 and test[i] < 3.0001 and v==2.50005:
                num_correct+=1
            elif test[i] >3.0000 and test[i] < 4.0001 and v==3.50005:
                num_correct+=1
            elif test[i] >4.0000 and test[i] < 5.0001 and v==4.50005:
                num_correct+=1
            elif test[i] >5.0000 and test[i] < 6.0001 and v==5.50005:
                num_correct+=1
            elif test[i] >6.0000 and test[i] < 7.0001 and v==6.50005:
                num_correct+=1
            elif test[i] >7.0000 and test[i] < 8.0001 and v==7.50005:
                num_correct+=1
            elif test[i] >8.0000 and test[i] < 9.0001 and v==8.50005:
                num_correct+=1
            elif test[i] >9.0000 and test[i] < 10.0001 and v==9.50005:
                num_correct+=1
            elif test[i] >10.0000 and test[i] < 15.0001 and v==12.50005:
                num_correct+=1
            elif test[i] >15.0000 and test[i] < 20.0001 and v==17.50005:
                num_correct+=1
            elif test[i] > 20.0000 and v==20.0000:
                num_correct+=1
        return num_correct/num_all
    else:
        return "Error: size of array is not same"
```

```python
data = pd.read_csv(data_path , header=0, usecols=[ 'salary',
       'education level',  'field', 'years of full time job',
       'years of learn programe', '[ASP .net]', '[Assembly]',
       '[Bash/Shell/PowerShell]', '[C]', '[C#]', '[C++]', '[Clojure]',
       '[Dart]', '[Elixir]', '[Erlang]', '[Go]', '[Groovy]', '[Haskell]',
       '[HTML/CSS]', '[Java]', '[JavaScript]', '[Julia]', '[Kotlin]', '[Lua]',
       '[Objective-C]', '[Ocaml]', '[Pascal]', '[Perl]', '[PHP]', '[Python]',
       '[R]', '[Ruby]', '[Rust]', '[Scala]', '[SQL]'])

y_all = data.iloc[:,0].values 
num_y_all = len(y_all)
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from math import sqrt
import matplotlib.pyplot as plt
import statsmodels.api as sm

def  MLR(data):
    y = data.iloc[:,0].values 
    X = data.iloc[:,1:]
    
    #set train and test
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
    regr = sm.OLS(y, sm.add_constant(X)).fit()
    print('size of data:',len(y),'\nsize of train:',len(y_train),'\nsize of test:',len(y_test))
    
    y_pred = regr.predict( sm.add_constant(X_test) )
    r2 = regr.rsquared
    
    print('r2_score: ',r2)
    r2_ratio = r2*(len(y)/num_y_all)
    rmse = sqrt(mean_squared_error(y_test,y_pred))
    nrmse = rmse/(max(y_test)-min(y_test))
    print('nrmse:',nrmse)
    acc = accuracy_range(y_test,y_pred.values)
    print('accuracy:',acc)
    plt.scatter(y_test,y_pred,color='b', s=10, alpha=0.8)
    plt.xlabel('True values')
    plt.ylabel('Predictions')
    plt.show()
    result_coef = regr.params
    print('Coefficients: \n', result_coef) 
        
    #confident interval
    cofinv = confidence_interval(data, rmse ,0.95)    
    print(regr.summary())
    
    return ( result_coef,r2_ratio, nrmse, acc,cofinv)
```

```python
import numpy as np
from scipy import stats

def mean_confidence_interval(data, confidence=0.95):
    a = 1.0 * np.array(data)
    n = len(a)
    m, se = np.mean(a), stats.sem(a)
    h = se * stats.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h, h

def confidence_interval(data, rmse, confidence=0.95):    
    y = data.iloc[:,0].values 
    X = data.iloc[:,1:].values
    
    k = len(data.columns)
    df = len(X) - k - 1                  # degress of freedom  n-k-1
    t = stats.t.ppf(confidence/2, df)   # t-critical value for 95%
    S = rmse                                #root mean square error
    n = len(X)

    X_ = np.array(X)
    X_tp = X.transpose()
    XtpX =  np.matmul(X_tp,X_)
    C = np.linalg.inv(XtpX)

    # x = np.array(X_test[0])
    # x_tp = x.transpose()
    # xtpCx = np.matmul(np.matmul(x_tp,C),x)

    # h = t*S*math.sqrt(((1/n)+xtpCx))

    # print(yhat[0]-h ,yhat[0]+h)
    return [t,S,n,C]
```

```python
def histogram_intersection(a, b):
    v = np.minimum(a, b).sum().round(decimals=1)
    return v
```
